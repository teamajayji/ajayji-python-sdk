```python
import asyncio
from ajayji import SDK
from ajayji.mcp_host import JupyterMCPHost

# 1. Initialize our MCP Host
mcp_host = JupyterMCPHost(name="NotebookServer", port=8000)

# 2. Define a custom Python function and register it as an MCP tool
@mcp_host.tool(name="calculate_complex_math", description="Multiplies a number by 42")
def calculate_complex_math(number: int) -> int:
    print(f"🐍 [Jupyter] Executing native python function with input: {number}")
    return number * 42

# 3. Start the FastMCP Server in the background
mcp_host.start()

async def run_test():
    # 4. Initialize the Ajayji SDK
    sdk = SDK()
    
    agent_id = "test_mcp_agent"
    
    # 5. Provision the Persona using the new SDK method we generated
    print("📡 [SDK] Provisioning Persona with MCP Server...")
    provision_response = await sdk.personas.create_persona_async(
        agent_id=agent_id,
        agent_name="MCP Demo Agent",
        model="llama-3-8b.gguf", # Change this to an actively downloaded model filename you have!
        mcp_servers=["http://127.0.0.1:8000/sse"],
        system_prompt="You have access to a tool called 'calculate_complex_math'. You must use it when asked to perform math."
    )
    print(f"✅ [SDK] Provision Response: {provision_response}")

    # 6. Invoke the Persona, which will trigger the Daemon to connect back to us
    print("🤖 [SDK] Invoking Persona with prompt...")
    invoke_response = await sdk.personas.invoke_persona_async(
        id=agent_id,
        request_body={
            "query": "Please use the calculate_complex_math tool on the number 10 and tell me the result."
        }
    )
    print(f"🎉 [SDK] Final Answer from Model: {invoke_response}")

# Execute the test
if __name__ == "__main__":
    asyncio.run(run_test())
```

```python
import asyncio
from ajayji import SDK
from ajayji.mcp_host import JupyterMCPHost

# 1. Initialize our MCP Host (Using 8001 to avoid 'address already in use')
mcp_host = JupyterMCPHost(name="NotebookServer", port=8002)

# 2. Define a custom Python function and register it as an MCP tool
@mcp_host.tool(name="calculate_complex_math", description="Multiplies a number by 42")
def calculate_complex_math(number: int) -> int:
    print(f"🐍 [Jupyter] Executing native python function with input: {number}")
    return number * 42

# 3. Start the FastMCP Server in the background
mcp_host.start()

async def run_test():
    # 4. Initialize the Ajayji SDK
    sdk = SDK()
    
    agent_id = "test_mcp_agent"
    
    # 5. Provision the Persona using the new SDK method we generated
    print("📡 [SDK] Provisioning Persona with MCP Server...")
    provision_response = await sdk.persona_orchestration.create_persona_async(
        agent_id=agent_id,
        agent_name="MCP Demo Agent",
        model="gemma-3-1b-it-Q5_K_M.gguf", # Change this to an actively downloaded model filename you have!
        mcp_servers=["http://127.0.0.1:8002/sse"], # Notice we match port 8001 here!
        system_prompt="You have access to a tool called 'calculate_complex_math'. You must use it when asked to perform math."
    )
    print(f"✅ [SDK] Provision Response: {provision_response}")

    # 6. Invoke the Persona, which will trigger the Daemon to connect back to us
    print("🤖 [SDK] Invoking Persona with prompt...")
    invoke_response = await sdk.persona_orchestration.invoke_persona_async(
        id=agent_id,
        request_body={
            "query": "Please use the calculate_complex_math tool on the number 10 and tell me the result."
        }
    )
    print(f"🎉 [SDK] Final Answer from Model: {invoke_response}")

# Execute the test directly in Jupyter!
await run_test()
```
