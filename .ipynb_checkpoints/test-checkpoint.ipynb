```python
import pandas as pd
import openapi
import os

client = openapi.SDK(server_url="http://localhost:14321")

# Optional: Print out what properties the SDK generated to be absolutely sure
# print("Available SDK Modules:", [m for m in dir(client) if not m.startswith('_')])

# 1. Create a dummy dataframe in memory
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [28, 34, 22],
    'role': ['Data Scientist', 'Engineer', 'Analyst']
})

temp_csv_path = os.path.abspath('temp_data.csv')
df.to_csv(temp_csv_path, index=False)

# 2. Use the correct Speakeasy generated module name for the tag
response = client.data_ingestion_and_tools.create_database(
    name="My Pandas Database",
    csv_path=temp_csv_path
)
print("Database Provision Response:", response)
```
