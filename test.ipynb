```python
import pandas as pd
import ajayji
import os

# We now import the correct 'ajayji' module!
client = ajayji.SDK(server_url="http://localhost:14321")

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [28, 34, 22],
    'role': ['Data Scientist', 'Engineer', 'Analyst']
})

temp_csv_path = os.path.abspath('temp_data.csv')
df.to_csv(temp_csv_path, index=False)

response = client.data_ingestion_and_tools.create_database(
    name="My Pandas Database",
    csv_path=temp_csv_path
)
print("Database Provision Response:", response)
```

```python
import urllib.request
import os
import ajayji

dummy_model_url = "https://github.com/onnx/models/raw/main/vision/classification/mnist/model/mnist-8.onnx"
model_path = os.path.abspath("mnist.onnx")

if not os.path.exists(model_path):
    print("Downloading dummy model...")
    urllib.request.urlretrieve(dummy_model_url, model_path)

client = ajayji.SDK(server_url="http://localhost:14321")

# Use the correct tag module generated by Speakeasy
response = client.model_management.convert_model(
    name="My Test CoreML Model",
    source_file_path=model_path,
    type="image",
    input_mapping="Input3",
    output_mapping="Plus214_Output_0"
)

print("Conversion Stream logs:")
print(response)
```

```python
import time
from ajayji import SDK

sdk = SDK()

print("1. Loading Embedding Model into RAM...")
# This calls the /run endpoint to load it into active memory!
sdk.memory_management.load_model(
    file_name="embeddinggemma-300m-Q4_0.gguf"
)

# Optional sleep to ensure the model is fully loaded before hitting the vector store
time.sleep(5) 

print("2. Creating Vector Database...")
# Now that it's loaded, we can run the ingestion!
res = sdk.data_ingestion_and_tools.create_vector_db(
    name="My Vector Database",
    folder_path="/Users/ramanyala/Downloads/code/fluttercode/test_docs", 
    embedding_model="embeddinggemma-300m-Q4_0.gguf"
)
# Fixed typo here!
print(res.status)

print("3. Querying Vector Database...")
query_res = sdk.data_ingestion_and_tools.execute_tool(
    tool_name="My Vector Database",
    args={
        "query": "What is the main topic of the document?"
    }
)
print("Search Results:", query_res)
```
